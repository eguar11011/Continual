```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pathlib import Path

# Cargar las matrices desde archivos CSV en el entorno
task0_path = Path("confmat_task0.csv")
task1_path = Path("confmat_task1.csv")
global_path = Path("confmat_global.csv")

# Cargar los datos
confmat_task0 = pd.read_csv(task0_path, index_col=0)
confmat_task1 = pd.read_csv(task1_path, index_col=0)
confmat_global = pd.read_csv(global_path, index_col=0)

# Función para graficar una matriz de confusión
def plot_confusion_matrix(confmat, title):
    plt.figure(figsize=(8, 6))
    sns.heatmap(confmat, annot=True, fmt='d', cmap='Blues', cbar=True)
    plt.title(title)
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.tight_layout()
    plt.show()

# Graficar las tres matrices
plot_confusion_matrix(confmat_task0, "Matriz de Confusión - Task 0")
plot_confusion_matrix(confmat_task1, "Matriz de Confusión - Task 1")
plot_confusion_matrix(confmat_global, "Matriz de Confusión - Global")
```
